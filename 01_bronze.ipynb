```python
from pyspark.sql.functions import current_timestamp
from tqdm import tqdm


files = [
    "brands",
    "categories",
    "products",
    "customers",
    "stores",
    "staffs",
    "orders",
    "order_items",
    "stocks",
]

for file in tqdm(files):
    path = f"dbfs:/Volumes/c_bikeshop/raw/raw_files/{file}.csv"
    df = (
        spark.read.option("header", True)
        .option("inferSchema", True)
        .csv(path)
        .withColumn("_ingest_ts", current_timestamp())
    )
    (
        df.write.mode("overwrite")
        .format("delta")
        .saveAsTable(f"c_bikeshop.bronze.{file}")
    )
    print(f"Loaded bronze.{file}")
```
